```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
import pandas_datareader as pdr
data = pdr.get_data_fred("HOUSTNSA", "1959-01-01", "2019-06-01")
```

```python
fig, ax = plt.subplots(1,1,figsize=(7,4))

ax.plot(data,label='Taxa',alpha=0.6)
ax.plot(data.rolling(12).mean(),label='média móvel 12M',color='r')

ax.set_title('crescimento mensal de habitações nos EUA')
ax.legend(fontsize=10,loc='upper left')

plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(data)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
if result[1] < 0.05:
  print('Série é estacionária')
else:
  print('Série não é estacionária')
```

```python
from statsmodels.tsa.stattools import kpss

result = kpss(data)
print('Kpss Statistic: %f' % result[0])
print('p-value: %f' % result[1])
if result[1] < 0.05:
  print('Série é estacionária')
else:
  print('Série não é estacionária')
```
